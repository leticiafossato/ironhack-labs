#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
fig,ax=plt.subplots()
y=x**2
ax.plot(x,y)
y=np.exp(x)
ax.plot(x,y)
plt.title('Normal Scale')
plt.show()

fig,ax=plt.subplots()
y=x**2
ax.set_xscale('log')
ax.set_yscale('log')
ax.plot(x,y)
y=np.exp(x)
ax.plot(x,y)
plt.title('Logarithmic Scale')
plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
fig,ax=plt.subplots()
y=x**2
ax.plot(x,y)
y=np.exp(x)
ax.plot(x,y)
plt.title('Normal Scale')
plt.legend(['$x^2$','$e^x$'])
plt.show()

fig,ax=plt.subplots()
y=x**2
ax.set_xscale('log')
ax.set_yscale('log')
ax.plot(x,y)
y=np.exp(x)
ax.plot(x,y)
plt.title('Logarithmic Scale')
plt.legend(['$x^2$','$e^x$'])
plt.show()
```
