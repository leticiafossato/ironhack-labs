# StackAPI

#### Import the necessary libraries here:

```python
# your code here
import requests
from stackapi import StackAPI
from datetime import datetime
import pandas as pd
```

#### Question 1: Find the questions and answers of last month.

```python
# your code here

SITE = StackAPI('stackoverflow')

# the time used in this API is in Unix epoch time!
# https://en.wikipedia.org/wiki/Unix_time


date_start = int(datetime.timestamp(datetime(2020,5,1)))
date_end = int(datetime.timestamp(datetime(2020,5,31)))

questions = SITE.fetch('questions', fromdate=date_start, todate=date_end)
answers = SITE.fetch('answers', fromdate=date_start, todate=date_end)
```

```python
questions
```

```python
questions_df = pd.DataFrame(questions['items'])
answers_df = pd.DataFrame(answers['items'])
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
SITE = StackAPI('stackoverflow')
date_start = int(datetime.timestamp(datetime(2020, 6, 21, 23, 59, 59)))
date_end = int(datetime.timestamp(datetime(2020, 6, 22, 23, 59, 59)))

questions_today = SITE.fetch('questions', fromdate=date_start, todate=date_end, tagged=['python'], sort='votes', min = 5, order = 'desc')
```

```python
questions_today_df = pd.DataFrame(questions_today['items'])
```

```python
questions_today_df
```
